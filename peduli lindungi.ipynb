```python
df.groupby(['nik_hashed']).agg(['count'])['user_covid19_status']
df_order=df[df['outlet_sub_category'].isin(['RESTORAN','SUPERMARKET','BANK'])].groupby(['checkin_date','outlet_sub_category']).agg(['count'])['user_covid19_status'].sort_values('checkin_date')
    
```

```python
df_order.pivot_table(values='count', index='checkin_date', columns='outlet_sub_category', aggfunc='first').to_csv('pola traffic.csv')
```
